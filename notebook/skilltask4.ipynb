```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("C:\\Users\\Apurva HG\\anaconda3\\share\\jupyter\\lab\\task4.xlsx")
df['Start_Time'] = pd.to_datetime(df['Start_Time'])
df['Hour'] = df['Start_Time'].dt.hour
df['Day'] = df['Start_Time'].dt.day_name()
df['Month'] = df['Start_Time'].dt.month_name()

df['Weather_Condition'] = df['Weather_Condition'].fillna('Unknown')
df['Severity'] = pd.to_numeric(df['Severity'], errors='coerce')

df.groupby('Weather_Condition')['Severity'].mean().sort_values().plot(kind='barh', figsize=(10,6), color='skyblue')
plt.title('Average Severity by Weather Condition')
plt.xlabel('Average Severity')
plt.ylabel('Weather Condition')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

features = ['Junction', 'Traffic_Signal', 'Bump', 'Crossing', 'Stop']
counts = df[features].sum()

plt.figure(figsize=(8,8))
counts.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Accidents Involving Road Features')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

severity_by_hour = df.groupby('Hour')['Severity'].mean()

plt.figure(figsize=(10,6))
severity_by_hour.plot(kind='line', marker='o', color='darkred')
plt.title('Average Severity by Hour')
plt.xlabel('Hour of Day')
plt.ylabel('Average Severity')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.scatter(df['Visibility(mi)'], df['Severity'], alpha=0.5, color='teal')
plt.title('Severity vs Visibility')
plt.xlabel('Visibility (miles)')
plt.ylabel('Severity')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,8))
sns.kdeplot(
    x=df['Start_Lng'], 
    y=df['Start_Lat'], 
    cmap="Reds", 
    shade=True, 
    bw_adjust=0.5
)
plt.title('Accident Hotspots')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.tight_layout()
plt.show()
```
